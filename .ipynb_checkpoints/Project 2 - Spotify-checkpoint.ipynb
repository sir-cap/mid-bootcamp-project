```python
def plot_distribution(spotify):
    for col in spotify.columns:
        stat, p = shapiro(spotify[col])

        if p > 0.05:
            print('The data is normal.')
        else:
            print('The data is likely not normal.')
            
    return spotify
```

```python
plot_distribution(spotify)
```

```python
from sklearn.preprocessing import Normalizer
transformer = Normalizer().fit(spotify)
spotify_normalized = transformer.transform(spotify)
spotify = pd.DataFrame(spotify_normalized)
```
